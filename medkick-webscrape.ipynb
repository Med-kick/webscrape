```python
def scroll_to_element(element):
    browser.execute_script("arguments[0].scrollIntoView();", element)

def click_element_with_offset(element):
    actions = ActionChains(browser)
    actions.move_to_element(element)
    actions.move_by_offset(5, 5)  # Adjust the offset as needed
    actions.click()
    actions.perform()
```

```python
#merge all csv to one csv file
# Step 1: Specify the folder path containing the CSV files
folder_path = '/Users/gulipalli/Desktop/medkick/data/'

# Step 2: Get the list of CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Step 3: Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Step 4: Merge the CSV files
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    merged_data = merged_data.append(df)

# Step 5: Write the merged data to a new CSV file
output_file_path = '/Users/gulipalli/Desktop/medkick/data/merged_data.csv'
merged_data.to_csv(output_file_path, index=False)

print("Merged CSV files successfully and saved to:", output_file_path)
```

```python
# Streamlit dashboard
import streamlit as st
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import plotly.express as px  # pip install plotly-express
import plotly.graph_objects as go

# Set page configuration
st.set_page_config(page_title="Nursing Dashboard", page_icon="\U0001F4C8", layout="wide")
with open('style.css') as f:
    st.markdown(f'<style>{f.read()}</style>', unsafe_allow_html=True)
# Function to read data from CSV file
def get_data_from_csv():
    df = pd.read_csv('/Users/gulipalli/Desktop/medkick/data/merged_data.csv')
    # Drop the specified columns
    columns_to_drop = ['Disposition', 'End Time (local)', 'Answer Time (local)', 'Start Time (local)']
    df = df.drop(columns=columns_to_drop)
    # Convert '--:--:--' to NaN
    df['Answer Time'] = np.where(df['Answer Time'] == '--:--:--', np.nan, df['Answer Time'])
    # Count the number of rows where 'Answer Time' was converted
    rows_converted = df['Answer Time'].isnull().sum()
    print("Number of rows converted:", rows_converted)
    # Convert the 'Start Time', 'Answer Time', and 'End Time' columns to datetime
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['Answer Time'] = pd.to_datetime(df['Answer Time'])
    df['End Time'] = pd.to_datetime(df['End Time'])
    # Extract the date and time components
    df['Date'] = df['Start Time'].dt.date
    df = df.sort_values('Date', ascending=True)
    return df
# Get the data from CSV
df = get_data_from_csv()
# Display the welcome message
#st.title("Welcome to the Nursing Dashboard! \U0001F469‍⚕️")
st.markdown('<h1 style="color: white;">Welcome to the Nursing Dashboard! \U0001F469‍⚕️</h1>', unsafe_allow_html=True)

# Sidebar filters
st.sidebar.header("Please Filter Here:")
direction_options = df["Direction"].unique()
nurse_options = df["From"].unique()
customer_options = df["To"].unique()
# Multiselect filters
direction = st.sidebar.multiselect("Select the direction:", options=direction_options)
nurse = st.sidebar.multiselect("Select the nurse:", options=nurse_options)
customer = st.sidebar.multiselect("Select the customer:", options=customer_options)
# Date range selection
min_date = min(df['Date'])
max_date = max(df['Date'])
start_date = st.sidebar.date_input("Select start date:", value=min_date, min_value=min_date, max_value=max_date)
end_date = st.sidebar.date_input("Select end date:", value=max_date, min_value=min_date, max_value=max_date)
selected_date = st.sidebar.date_input("Select a date:", value=max_date, min_value=min_date, max_value=max_date)
# Filter the DataFrame based on the selected filters
filtered_df = df.copy()
if direction:
    filtered_df = filtered_df[filtered_df["Direction"].isin(direction)]
if nurse:
    filtered_df = filtered_df[filtered_df["From"].isin(nurse)]
if customer:
    filtered_df = filtered_df[filtered_df["To"].isin(customer)]
filtered_df = filtered_df[(filtered_df['Date'] >= start_date) &(filtered_df['Date'] <= end_date)]


# Calculate average durations
average_duration = filtered_df.groupby('Direction')['Duration'].mean()
average_outbound_duration = filtered_df[filtered_df['Direction'] == 'OUTBOUND']['Duration'].mean()
average_inbound_duration = filtered_df[filtered_df['Direction'] == 'INBOUND']['Duration'].mean()

# Group the data by date, nurse, and direction for inbound and outbound calls
inbound_calls_nurse = filtered_df[filtered_df['Direction'] == 'INBOUND'].groupby(['Date', 'To']).size().reset_index(name='Inbound Count')
outbound_calls_nurse = filtered_df[filtered_df['Direction'] == 'OUTBOUND'].groupby(['Date', 'From']).size().reset_index(name='Outbound Count')
nurse_ibcount = inbound_calls_nurse[inbound_calls_nurse['Date'] == selected_date]
nurse_ibcount = nurse_ibcount['Inbound Count'].sum()
nurse_obcount = outbound_calls_nurse[outbound_calls_nurse['Date'] == selected_date]
nurse_obcount = nurse_obcount['Outbound Count'].sum()

# Group the data by date and 'From' for inbound calls and 'To' for outbound calls to identify same customer
inbound_calls_customer = filtered_df[filtered_df['Direction'] == 'INBOUND'].groupby(['Date', 'From']).size().reset_index(name='Inbound Count')
outbound_calls_customer = filtered_df[filtered_df['Direction'] == 'OUTBOUND'].groupby(['Date', 'To']).size().reset_index(name='Outbound Count')
total_unique_customers = len(outbound_calls_customer[outbound_calls_customer['Date'] == selected_date])
customer_obcount = outbound_calls_customer[outbound_calls_customer['Date'] == selected_date]
customer_obcount = customer_obcount['Outbound Count'].sum()


# Filter the DataFrame for after hours inbound calls
after_hours_inbound_calls = filtered_df[(filtered_df['Direction'] == 'INBOUND') & ((filtered_df['Start Time'].dt.hour >= 18) | (filtered_df['Start Time'].dt.hour < 8))]
# Filter the DataFrame for weekend inbound calls
weekend_inbound_calls = filtered_df[(filtered_df['Direction'] == 'INBOUND') & (filtered_df['Start Time'].dt.weekday >= 5)]
# Filter the DataFrame for after hours outbound calls
after_hours_outbound_calls = filtered_df[(filtered_df['Direction'] == 'OUTBOUND') & ((filtered_df['Start Time'].dt.hour >= 18) | (filtered_df['Start Time'].dt.hour < 8))]
# Filter the DataFrame for weekend outbound calls
weekend_outbound_calls = filtered_df[(filtered_df['Direction'] == 'OUTBOUND') & (filtered_df['Start Time'].dt.weekday >= 5)]
# Group the data by date, nurse, and direction for after hours and weekend inbound calls
after_hours_inbound_calls_per_nurse = after_hours_inbound_calls.groupby(['To']).size().reset_index(name='After Hours Inbound Count')
after_hours_ib = after_hours_inbound_calls_per_nurse['After Hours Inbound Count'].sum()
weekend_inbound_calls_per_nurse = weekend_inbound_calls.groupby([ 'To']).size().reset_index(name='Weekend Inbound Count')
weekend_ib = weekend_inbound_calls_per_nurse['Weekend Inbound Count'].sum()
# Group the data by date, nurse, and direction for after hours and weekend outbound calls
after_hours_outbound_calls_per_nurse = after_hours_outbound_calls.groupby(['From']).size().reset_index(name='After Hours Outbound Count')
after_hours_ob = after_hours_outbound_calls_per_nurse['After Hours Outbound Count'].sum()
weekend_outbound_calls_per_nurse = weekend_outbound_calls.groupby(['From']).size().reset_index(name='Weekend Outbound Count')
weekend_ob = weekend_outbound_calls_per_nurse['Weekend Outbound Count'].sum()

# Create a creative visualization for the key insights
fig_insights = go.Figure()

# Add average call durations
fig_insights.add_trace(go.Indicator(
    mode="number",
    value=average_outbound_duration,
    title={"text": "<br>Avg Outbound<br>Call Duration"},
    delta={"reference": average_inbound_duration, "relative": True},
    domain={'row': 0, 'column': 0},
    number={"font": {"size": 50}}
))

fig_insights.add_trace(go.Indicator(
    mode="number",
    value=average_inbound_duration,
    title={"text": "<br>Avg Inbound<br>Call Duration"},
    delta={"reference": average_outbound_duration, "relative": True},
    domain={'row': 0, 'column': 1},
    number={"font": {"size": 50}}
))

# Add total in-calls and out-calls
fig_insights.add_trace(go.Indicator(
    mode="number",
    value=nurse_ibcount,
    title={"text": "<br>Total In-Calls<br>To Nurse"},
    domain={'row': 1, 'column': 0},
    number={"font": {"size": 50}}
))

fig_insights.add_trace(go.Indicator(
    mode="number",
    value=nurse_obcount,
    title={"text": "<br>Total Out-Calls<br>From Nurse"},
    domain={'row': 1, 'column': 1},
    number={"font": {"size": 50}}
))

# Add total unique customers and out-calls to customers
fig_insights.add_trace(go.Indicator(
    mode="number",
    value=total_unique_customers,
    title={"text": "<br>Total Unique<br>Customers Called To"},
    domain={'row': 0, 'column': 2},
    number={"font": {"size": 50}}
))

fig_insights.add_trace(go.Indicator(
    mode="number",
    value=customer_obcount,
    title={"text": "<br>Total Out-Calls<br>To Customers"},
    domain={'row': 0, 'column': 3},
    number={"font": {"size": 50}}
))

# Add after hours and weekend call counts
fig_insights.add_trace(go.Indicator(
    mode="number",
    value=after_hours_ib,
    title={"text": "<br>After Hours<br>In-Calls Count"},
    domain={'row': 1, 'column': 2},
    number={"font": {"size": 50}}
))

fig_insights.add_trace(go.Indicator(
    mode="number",
    value=weekend_ib,
    title={"text": "<br>Weekend<br>In-Calls Count"},
    domain={'row': 1, 'column': 3},
    number={"font": {"size": 50}}
))

fig_insights.add_trace(go.Indicator(
    mode="number",
    value=after_hours_ob,
    title={"text": "<br>After Hours<br>Out-Calls Count"},
    domain={'row': 0, 'column': 4},
    number={"font": {"size": 50}}
))

fig_insights.add_trace(go.Indicator(
    mode="number",
    value=weekend_ob,
    title={"text": "<br>Weekend<br>Out-Calls Count"},
    domain={'row': 1, 'column': 4},
    number={"font": {"size": 50}}
))

# Update layout
fig_insights.update_layout(
    title="<b><span style='font-size: 32px; animation: rainbow 2s ease-in-out infinite; color: #00A3E0;'>Overview</span></b>",
    grid={'rows': 2, 'columns': 5, 'pattern': "independent"},
    template="plotly_white"
)


# Display the creative visualization
st.plotly_chart(fig_insights, use_container_width=True)

st.markdown('<h1 style="color: #00A3E0;">\U0001F4CA DATA summary</h1>', unsafe_allow_html=True)

#st.title(":bar_chart: DATA summary")
# Display the filtered DataFrame
st.dataframe(filtered_df)

# Create bar chart for average durations
fig_avg_duration = px.bar(
    x=average_duration.index,
    y=average_duration.values,
    title="<b>Average Call Duration</b>",
    color_discrete_sequence=["#a0db8e"] * len(average_duration),
    template="plotly_white"
)

# Create a function to filter the DataFrame by date and plot the inbound and outbound call counts for each nurse
def plot_incalls_to_nurse(selected_date):
    # Filter the DataFrame by the selected date
    filtered_calls = inbound_calls_nurse[inbound_calls_nurse['Date'] == selected_date]
    # Create a bar chart for the inbound call count on the selected date
    fig_inbound_calls = px.bar(
        filtered_calls,
        x='To',
        y='Inbound Count',
        color='To',
        title=f"<b>Inbound Call Count for {selected_date}</b>",
        template='plotly_white'
    )
    return fig_inbound_calls

def plot_outcalls_from_nurse(selected_date):
    # Filter the DataFrame by the selected date
    filtered_calls = outbound_calls_nurse[outbound_calls_nurse['Date'] == selected_date]
    # Create a bar chart for the inbound call count on the selected date
    fig_outbound_calls = px.bar(
        filtered_calls,
        x='From',
        y='Outbound Count',
        color='From',
        title=f"<b>Outbound Call Count for {selected_date}</b>",
        template='plotly_white'
    )
    return fig_outbound_calls

# Call the function to plot the inbound and outbound call counts for the selected date
fig_icalls = plot_incalls_to_nurse(selected_date)
fig_ocalls = plot_outcalls_from_nurse(selected_date)

# Create a function to plot the bar chart for the count of inbound calls from the same number per day
def plot_inbound_calls_from_customer(selected_date):
    filtered_inbound_calls = inbound_calls_customer[inbound_calls_customer['Date'] == selected_date]
    fig_inbound_calls_bar = px.bar(
        filtered_inbound_calls,
        x='From',
        y='Inbound Count',
        title=f"<b>Count of Inbound Calls from the Same Number on {selected_date}</b>",
        template='plotly_white'
    )
    return fig_inbound_calls_bar

def plot_outbound_calls_to_customer(selected_date):
    filtered_outbound_calls = outbound_calls_customer[outbound_calls_customer['Date'] == selected_date]
    fig_outbound_calls_bar = px.bar(
        filtered_outbound_calls,
        x='To',
        y='Outbound Count',
        title=f"<b>Count of Outbound Calls to the Same Number on {selected_date}</b>",
        template='plotly_white'
    )
    return fig_outbound_calls_bar
# Call the functions to get the figures
fig_inbound_calls_bar = plot_inbound_calls_from_customer(selected_date)
fig_outbound_calls_bar = plot_outbound_calls_to_customer(selected_date)




# Create a bar chart for after hours inbound calls per nurse per day
fig_after_hours_inbound_calls_per_nurse = px.bar(
    after_hours_inbound_calls_per_nurse,
    x='To',
    y='After Hours Inbound Count',
    title='<b>After Hours Inbound Calls per Nurse</b>',
    template='plotly_white'
)
# Create a bar chart for weekend inbound calls per nurse per day
fig_weekend_inbound_calls_per_nurse = px.bar(
    weekend_inbound_calls_per_nurse,
    x='To',
    y='Weekend Inbound Count',
    title='<b>Weekend Inbound Calls per Nurse</b>',
    template='plotly_white'
)
# Create a bar chart for after hours outbound calls per nurse per day
fig_after_hours_outbound_calls_per_nurse = px.bar(
    after_hours_outbound_calls_per_nurse,
    x='From',
    y='After Hours Outbound Count',
    title='<b>After Hours Outbound Calls per Nurse </b>',
    template='plotly_white'
)
# Create a bar chart for weekend outbound calls per nurse per day
fig_weekend_outbound_calls_per_nurse = px.bar(
    weekend_outbound_calls_per_nurse,
    x='From',
    y='Weekend Outbound Count',
    title='<b>Weekend Outbound Calls per Nurse</b>',
    template='plotly_white'
)

inbound_filtered_df = filtered_df[filtered_df['Direction'] == 'INBOUND']
outbound_filtered_df = filtered_df[filtered_df['Direction'] == 'OUTBOUND']
inbound_filtered_df['Time between calls - with no dials'] = inbound_filtered_df.groupby('To')['Start Time'].shift(-1) - inbound_filtered_df['End Time']
inbound_filtered_df['Time between calls - with no dials'] = inbound_filtered_df['Time between calls - with no dials'].dt.total_seconds() / 60
# Calculate the time between calls (with no dials) for each nurse in outbound calls
outbound_filtered_df['Time between calls - with no dials'] = outbound_filtered_df.groupby('From')['Start Time'].shift(-1) - outbound_filtered_df['End Time']
outbound_filtered_df['Time between calls - with no dials'] = outbound_filtered_df['Time between calls - with no dials'].dt.total_seconds() / 60
# Calculate the average time between consecutive calls for each nurse
inbound_avg_time = inbound_filtered_df.groupby('To')['Time between calls - with no dials'].mean().reset_index()
outbound_avg_time = outbound_filtered_df.groupby('From')['Time between calls - with no dials'].mean().reset_index()

# Plot the average time between consecutive calls for each nurse in inbound and outbound calls
fig_avg_time_inbound = px.bar(
    inbound_avg_time,
    x='To',
    y='Time between calls - with no dials',
    title='<b>Average Time Between Consecutive Calls (with no dials) - Inbound</b>',
    template='plotly_white'
)

fig_avg_time_outbound = px.bar(
    outbound_avg_time,
    x='From',
    y='Time between calls - with no dials',
    title='<b>Average Time Between Consecutive Calls (with no dials) - Outbound</b>',
    template='plotly_white'
)



# Display the charts in full-width layout
with st.beta_container():
    st.plotly_chart(fig_avg_duration, use_container_width=True)
    st.plotly_chart(fig_icalls, use_container_width=True)
    st.plotly_chart(fig_ocalls, use_container_width=True)
    st.plotly_chart(fig_inbound_calls_bar, use_container_width=True)
    st.plotly_chart(fig_outbound_calls_bar, use_container_width=True)
    st.plotly_chart(fig_after_hours_inbound_calls_per_nurse,  use_container_width=True)
    st.plotly_chart(fig_weekend_inbound_calls_per_nurse, use_container_width=True)
    st.plotly_chart(fig_after_hours_outbound_calls_per_nurse, use_container_width=True)
    st.plotly_chart(fig_weekend_outbound_calls_per_nurse, use_container_width=True)
    st.plotly_chart(fig_avg_time_inbound, use_container_width=True)
    st.plotly_chart(fig_avg_time_outbound, use_container_width=True)
```
